```python
def rw(steps):
    x = np.zeros(steps) 
    y = np.zeros(steps) 
  
    for i in range(1, steps): 
        val = rd.randint(1, 4) 
        if val == 1: 
            x[i] = x[i - 1] + 1
            y[i] = y[i - 1] 
        elif val == 2: 
            x[i] = x[i - 1] - 1
            y[i] = y[i - 1] 
        elif val == 3: 
            x[i] = x[i - 1] 
            y[i] = y[i - 1] + 1
        else: 
            x[i] = x[i - 1] 
            y[i] = y[i - 1] - 1

    return [x,y]    
```

```python
steps = 20
x, y = rw(steps) 
  
plt.title("Random Walk ($n = " + str(steps) + "$ steps)") 
plt.plot(x, y) 
plt.plot(x[0],y[0],'go')
plt.plot(x[-1],y[-1],'rs')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["animation.html"] = "html5"
import matplotlib.animation
import numpy as np

fig, ax = plt.subplots(figsize = (5,5))

# ax.grid()
ax.axis('square')

ax.set_xlim([min(x)-1,max(x)+1])
ax.set_ylim([min(y)-1,max(y)+1])
ax.scatter(x[0], y[0], c = "g")
def animate(i):
    ax.cla()
    # ax.grid()
    ax.set_xlim([min(x)-1, max(x)+1])
    ax.set_ylim([min(y)-1, max(y)+1])
    
    ax.scatter(x[:i+1], y[:i+1], c = "blue")
    ax.plot(x[:i+1], y[:i+1], "-", c = "blue")
    
    ax.scatter(x[0], y[0], s = 100 ,c = "g")
    ax.scatter(x[i], y[i], s = 100, c = "black")
    if i == len(x)-1:
        ax.scatter(x[-1], y[-1], s = 100 ,c = "r")
    return ax,
ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(x))
```
