```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["animation.html"] = "html5"
import numpy as np

import random as rd
import numpy as np
from statsmodels.distributions.empirical_distribution import ECDF
import pandas as pd
import networkx as nx
import seaborn as sns

sns.set()
```

```python
def rw_2(steps, a = 1):
    pi = np.pi
    args = np.random.uniform(0, 2*pi, steps)
    x = np.cumsum(np.sin(args))
    y = np.cumsum(np.cos(args))
    return [x,y]
```

```python
steps = 1000
x, y = rw_2(steps) 
plt.title("Random Walk ($n = " + str(steps) + "$ steps)") 
plt.plot(x, y) 
plt.plot(x[0], y[0],'go')
plt.plot(x[-1],y[-1],'rs')
```

```python
N = 10000
A_N = np.zeros(N)
B_N = np.zeros(N)

for i in range(N):
    x, y = rw_2(steps)
    A_N[i] = np.sum(x>0)/steps
    B_N[i] = np.sum((x>0) & (y>0))/steps
    
print(f"A_n = {np.mean(A_N)}\nB_n = {np.mean(B_N)}")
plt.boxplot(A_N)
plt.ylabel("$A_N$")
plt.show()
plt.boxplot(B_N)
plt.ylabel("$B_N$")

plt.show()
```

```python
plt.hist(A_N, density=True, bins = 50)
plt.ylabel("$Density$")
plt.xlabel("$A_N$")
plt.show()
```

```python
A_N_ECDF = ECDF(A_N)
plt.plot(np.linspace(0,1,100), A_N_ECDF(np.linspace(0,1,100)))
plt.ylabel("$P$")
plt.xlabel("$A_N$")
plt.show()
```

```python
plt.hist(B_N, density=True, bins = 50)
plt.ylabel("$Density$")
plt.xlabel("$B_N$")
plt.show()
```

```python
B_N_ECDF = ECDF(B_N)
plt.plot(np.linspace(0,1,100), B_N_ECDF(np.linspace(0,1,100)))
plt.ylabel("$F_{B_N}$")
plt.xlabel("$B_N$")
plt.show()
```
